```python
from scipy.stats import beta
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.linspace(0,1,100)


fig, ax = plt.subplots(figsize=(10,8))
params = [0.1,2.0] # [a,b]
ax.plot(xs, beta(params[0],params[1]).pdf(xs), linewidth=3, label='$p_{10}\sim B(%0.1f,%0.1f)$'%(params[0],params[1]))

params = [0.5,2.0]
ax.plot(xs, beta(params[0],params[1]).pdf(xs), linewidth=3, label='$p_{01}\sim B(%0.1f,%0.1f)$'%(params[0],params[1]))

params = [0.8,2.0]
ax.plot(xs, beta(params[0],params[1]).pdf(xs), linewidth=3, label='$\\alpha\sim B(%0.1f,%0.1f)$'%(params[0],params[1]))
ax.plot([0,1],[0,0],'--k')

ax.set_xlim([0,1])
ax.legend(fontsize=16)
ax.set_xlabel('Value', fontsize=16)
ax.set_ylabel('Density', fontsize=16)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.title('Beta Prior Densities',fontsize=18)

plt.savefig('PriorDensities.png')
plt.show()
```
